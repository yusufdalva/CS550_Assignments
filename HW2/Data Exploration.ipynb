```python
DATA_FILE_NAMES = ["train1", "test1", "train2", "test2"]
BASE_URL = "http://www.cs.bilkent.edu.tr/~gunduz/teaching/cs550/documents"
FILE_URLS = [os.path.join(BASE_URL, file_name) for file_name in DATA_FILE_NAMES]

file_paths = download_data(DATA_FILE_NAMES, FILE_URLS)
print("File Locations for the first dataset")
print("Training data: {}".format(file_paths[0]))
print("Test data: {}".format(file_paths[1]))
print("File Locations for the second dataset")
print("Training data: {}".format(file_paths[2]))
print("Test data: {}".format(file_paths[3]))
```

```python
dataset_1 = Dataset(file_paths[0], file_paths[1])
dataset_2 = Dataset(file_paths[2], file_paths[3])
```

```python
# This function assumes that the input and outout are both 1 dimensional
def plot_data(train_data, test_data, title):
    assert type(title) == str
    train_features = train_data[:, 0]
    train_output = train_data[:, 1]
    test_features = test_data[:, 0]
    test_output = test_data[:, 1]
    
    fig, axes = plt.subplots(1,2, figsize=(16,6))
    fig.suptitle(title, fontsize=20)
    axes[0].scatter(x=train_features, y=train_output, c="blue")
    axes[0].set_title("Training Data", fontsize=14)
    axes[0].set_xlabel("x", fontsize=12)
    axes[0].set_ylabel("y", fontsize=12)
    axes[1].scatter(x=test_features, y=test_output, c="green")
    axes[1].set_title("Test data", fontsize=14)
    axes[1].set_xlabel("x", fontsize=12)
    axes[1].set_ylabel("y", fontsize=12)
```

```python
plot_data(dataset_1.train_data, dataset_1.test_data, "Data Distribution for Dataset 1")
```

```python
plot_data(dataset_2.train_data, dataset_2.test_data, "Data Distribution for Dataset 2")
```
