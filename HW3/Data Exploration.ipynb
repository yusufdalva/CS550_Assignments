## Loading the data

## Visualizing Pixel Density
To observe the pixel distribution on the image to be clustered, a 3D scatter plot has been used. As it can be seen from the plotted image, the red color is included in different pixel intensity values and is significant in the image. Due to this fact, the initial assumption about the assignment would be that the learned clusters would include dominant red color component. This assumption is made on the expectation that clusters can be learned successfully.

```python
def plot_3d_scatter(data):
    fig = plt.figure(figsize=(10,10))
    ax = fig.add_subplot(111, projection="3d")
    red_pixels = data[:, :, 0].flatten()
    green_pixels = data[:, :, 1].flatten()
    blue_pixels = data[:, :, 2].flatten()
    ax.scatter(red_pixels, green_pixels, blue_pixels, marker='o')
    ax.set_xlabel("Red Pixel Values")
    ax.set_ylabel("Green Pixel Values")
    ax.set_zlabel("Blue Pixel Values")
plot_3d_scatter(data)
```

## Visualizing the Image
In order to observe the data image in order to make a prediction on how should a clustered image should look, the following data visualization function is provided. The image to be clustered is given as the output of the next cell.

```python
def visualize_image(data):
    data = np.array(data, dtype=np.int_)
    fig = plt.figure(figsize=(10, 6))
    ax = fig.add_subplot(1,1,1)
    ax.set_title("Image to be clustered", fontsize=20)
    ax.imshow(data)
visualize_image(data)
```

## Visualizing RGB Color Intensities
In order to verify the assumption about the distribution of the pixel intensities, the color channel matrices for red, green and blue color are visualized in the following cell via the function provided. Interpreting this result, it is obvious that the red color dominates the image and green and blue colors are intense only on specific points (in white colored flower as an example). Thus the expectation of having cluster centered around a color which is a tone of red is still valid before perfoming any clustering operation. The visualization of RGB color channels are provided below.

```python
def plot_color_dist(data):
    red_pixels = data[:, :, 0]
    green_pixels = data[:, :, 1]
    blue_pixels = data[:, :, 2]
    fig, axes = plt.subplots(1, 3, figsize=(30, 8))
    axes[0].imshow(red_pixels, cmap="Reds")
    axes[0].set_title("Color Intensities (Red)", fontsize=20)
    axes[1].imshow(green_pixels, cmap="Greens")
    axes[1].set_title("Color Intensities (Green)", fontsize=20)
    axes[2].imshow(blue_pixels, cmap="Blues")
    axes[2].set_title("Color Intensities(Blue)", fontsize=20)
    fig.suptitle("Color Intensity Values of RGB Colors", fontsize=30)
plot_color_dist(data)
```
